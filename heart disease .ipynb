```python
df=pd.read_csv("heart.csv")
df.head()
```

```python
x=df.drop("HeartDisease",axis=1)
y=df["HeartDisease"]
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=100, random_state=42)# we can also use xgboost here in replacement of RandomForestClassifier
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
y_proba=model.predict_proba(x_test)
heart_disease_chances=y_proba[:,1]
heart_disease_chances_percent=(heart_disease_chances*100).round(0)
for i, chance in enumerate(heart_disease_chances_percent,start=1):
  print(f"patient {i} has chance of {int(chance)}% heart disease")
```
